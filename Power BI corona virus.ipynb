```python
import pandas as pd
```

```python
raw_data  = pd.read_csv('Data/owid-covid-data.csv')
raw_data.head()
```

```python
raw_data.columns
```

```python
raw_data=raw_data.drop(['new_cases_smoothed','new_deaths_smoothed', 'total_cases_per_million',
       'new_cases_per_million', 'new_cases_smoothed_per_million',
       'total_deaths_per_million', 'new_deaths_per_million',
       'new_deaths_smoothed_per_million', 'icu_patients',
       'icu_patients_per_million', 'hosp_patients',
       'hosp_patients_per_million', 'weekly_icu_admissions',
       'weekly_icu_admissions_per_million', 'weekly_hosp_admissions',
       'weekly_hosp_admissions_per_million', 'new_tests', 'total_tests',
       'total_tests_per_thousand', 'new_tests_per_thousand',
       'new_tests_smoothed', 'new_tests_smoothed_per_thousand',
       'positive_rate', 'tests_per_case', 'tests_units', 'total_vaccinations',
       'people_vaccinated', 'people_fully_vaccinated', 'total_boosters',
       'new_vaccinations', 'new_vaccinations_smoothed',
       'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred',
       'people_fully_vaccinated_per_hundred', 'total_boosters_per_hundred',
       'new_vaccinations_smoothed_per_million',
       'new_people_vaccinated_smoothed',
       'new_people_vaccinated_smoothed_per_hundred', 'female_smokers',
       'male_smokers', 
       'excess_mortality_cumulative_absolute', 'excess_mortality_cumulative',
       'excess_mortality', 'excess_mortality_cumulative_per_million'],axis=1)
raw_data.columns
```

```python
raw_data.head()
```

```python
raw_data.isnull().sum()
```

```python
raw_data['continent'].fillna(raw_data['location'],inplace=True)
raw_data['total_cases'].fillna(0,inplace=True)
raw_data['new_cases'].fillna(0,inplace=True)
raw_data['total_deaths'].fillna(0,inplace=True)
raw_data['new_deaths'].fillna(0,inplace=True)
raw_data['reproduction_rate'].fillna(round(raw_data['reproduction_rate'].mean(),2),inplace=True)
raw_data['stringency_index'].fillna(round(raw_data['stringency_index'].mean(),2),inplace=True)
raw_data['population'].fillna(round(raw_data['population'].mean(),2),inplace=True)
raw_data['population_density'].fillna(round(raw_data['population_density'].mean(),2),inplace=True)
raw_data['median_age'].fillna(round(raw_data['median_age'].median(),2),inplace=True)
raw_data['aged_65_older'].fillna(round(raw_data['aged_65_older'].mean(),2),inplace=True)
raw_data['aged_70_older'].fillna(round(raw_data['aged_70_older'].mean(),2),inplace=True)
raw_data['gdp_per_capita'].fillna(round(raw_data['gdp_per_capita'].mean(),2),inplace=True)
raw_data['extreme_poverty'].fillna(round(raw_data['extreme_poverty'].mean(),2),inplace=True)
raw_data['cardiovasc_death_rate'].fillna(round(raw_data['cardiovasc_death_rate'].mean(),2),inplace=True)
raw_data['diabetes_prevalence'].fillna(round(raw_data['diabetes_prevalence'].mean(),2),inplace=True)
raw_data['handwashing_facilities'].fillna(round(raw_data['handwashing_facilities'].mean(),2),inplace=True)
raw_data['hospital_beds_per_thousand'].fillna(round(raw_data['hospital_beds_per_thousand'].mean(),2),inplace=True)
raw_data['life_expectancy'].fillna(round(raw_data['life_expectancy'].mean(),2),inplace=True)
raw_data['human_development_index'].fillna(round(raw_data['human_development_index'].mean(),2),inplace=True)
```

```python
raw_data.isnull().sum()
```

```python
raw_data['date'] = pd.to_datetime(raw_data['date'])
raw_data['Month-Year'] = raw_data['date'].dt.strftime('%b-%Y')
```

```python
raw_data.to_csv('Data/CoronaVirus PowerBI Raw')
```
